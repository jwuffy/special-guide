**Data Cleaning:** Data set includes multiple entries on different dates for the same counties. It is necessary to remove duplicate counties. I will choose to keep the most recent entry as the unique row.

```python
clean_data = data.copy()
clean_data = data.drop_duplicates(["County", "State"], keep = "last")
clean_data
```

```python
# Question 1

#top 10 by Vehicle
top10bV = clean_data.loc[ :, ["County", "State", "Electric Vehicle (EV) Total"]]
top10bV.sort_values("Electric Vehicle (EV) Total" , axis=0, ascending=False, inplace = True)

top10bV = top10bV.head(10)
top10bV
```

```python
# Question 2:

#top 10 by Population
top10bP = clean_data.loc[ :, ["County", "State", "Percent Electric Vehicles"]]
top10bP.sort_values("Percent Electric Vehicles" , axis=0, ascending=False, inplace = True)

top10bP = top10bP.head(10)
top10bP
```

```python
# Question 3:
BEV = clean_data['Battery Electric Vehicles (BEVs)']
PHEV = clean_data['Plug-In Hybrid Electric Vehicles (PHEVs)']
total_BEV = BEV.sum()
total_PHEV = PHEV.sum()

#ratio is BEV : PHEV
ratio = total_BEV / total_PHEV

ratio
```
